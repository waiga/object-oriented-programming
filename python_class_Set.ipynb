# CLASS

```python
class Set:
  '''리스트 값을 받아서 set와 비슷한 형태로 반환받도록 임의로 만든 클래스'''

  def __init__(self, list_data=None):
    self.new_list = []
    self.count = 0
    if list_data:
      list_data.sort()
      for i in list_data:
        if i not in self.new_list:
          self.new_list.append(i)
          self.count += 1

  def add(self, elem):
    if elem not in self.new_list:
      self.new_list.append(elem)
      self.count += 1

  def discard(self, elem):
    if elem in self.new_list:
      self.new_list.remove(elem)
      self.count -= 1

  def clear(self):
    self.new_list = []
    self.count = 0

  def __len__(self):
    return self.count

  def __str__(self):
    string_type = '{'+', '.join([str(item) for item in self.new_list])+'}'
    return string_type

  def __contains__(self, elem):
    if elem in self.new_list:
      return True
    else:
      return False
  
  # self <= other: self가 other의 부분 집합이면 참 반환, 아니면 거짓 반환
  def __le__(self, other):
    '''self <= other'''
    self.check_list = []
    for i in self.new_list:
      if i in other.new_list:
        self.check_list.append(i)
    self.check_list.sort()
    if self.check_list == self.new_list:
      return True
    else:
      return False
  
  # self >= other: other가 self의 부분 집합이면 참, 아니면 거짓 반환
  def __ge__(self, other):
    '''self >= other'''
    self.check_list = []
    for i in other.new_list:
      if i in self.new_list:
        self.check_list.append(i)
    self.check_list.sort()
    if self.check_list == other.new_list:
      return True
    else:
      return False

  # self | other : self와 other 원소 모두 포함하는 합집합 set 반환
  def __or__(self, other):
    check_list = []
    for i in self.new_list:
      check_list.append(i)
    for j in other.new_list:
      if j not in check_list:
        check_list.append(j)
    check_list.sort()
    return Set(check_list) # 이 부분 또다시 자기 자신으로 호출해서 객체 형태로 만듦


  # self & other : self와 other 가 공통으로 포함하는 원소를 포함하는 교집합 Set 반환
  def __and__(self, other):
    check_list = []
    for i in self.new_list:
      if i in other.new_list:
        check_list.append(i)
    return Set(check_list)


  # self - other: self의 원소 중 other에 없는 원소만을 포함하는 차집합 set 반환
  def __sub__(self, other):
    check_list = []
    for item in self.new_list:
      if item not in other.new_list:
        check_list.append(item)
    return Set(check_list)

  # |=
  def __ior__(self, other):
    self.check_list = []
    for i in self.new_list:
      self.check_list.append(i)
    for j in other.new_list:
      if j not in self.check_list:
        self.check_list.append(j)
    self.check_list.sort()
    self.new_list = self.check_list
    return self

  # &=
  def __iand__(self, other):
    self.check_list = []
    for i in self.new_list:
      if i in other.new_list:
        self.check_list.append(i)
    self.new_list = self.check_list
    return self

  # -=
  def __isub__(self, other):
    self.check_list = []
    for item in self.new_list:
      if item not in other.new_list:
        self.check_list.append(item)
    self.new_list = self.check_list
    return self
```

# TEST

```python
a = Set([1, 2, 3, 4])
b = Set([1, 2, 3, 4])

print(a)
print(b)
print('-'*100)

a.discard(4)
b.discard(1)
print(a)
print(b)
print('-'*100)

print(len(a))
print(1 in a)
print(1 in b)
print('-'*100)

print(a | b)
print(a & b)
print(a - b)
print('-'*100)

print(a <= b)
print(a <= a | b)
print(a >= b)
print(a >= a & b)
print('-'*100)

b.clear()
print(b)
print('-'*100)

a = Set([1, 2, 3])
b = Set([3, 4])
address_a = id(a)
a |= b
print(a)
print(address_a == id(a))
print('-'*100)

a = Set([1, 2, 3])
b = Set([3, 4])
address_a = id(a)
a &= b
print(a)
print(address_a == id(a))
print('-'*100)

a = Set([1, 2, 3])
b = Set([3, 4])
address_a = id(a)
a -= b
print(a)
print(address_a == id(a))
print('-'*100)
```
